SISTEMA RECOMENDACION MIGRACION NETA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Mig_neta= pd.read_csv('Migracion Neta.csv')
```

```python
Mig_neta.head(10)
```

```python
Mig_neta.info()
```

```python
# Lista de columnas a eliminar
columns_to_drop = ['Indicator Name', 'Indicator Code']

# Eliminar las columnas
Mig_neta = Mig_neta.drop(columns=columns_to_drop)
```

```python
print(Mig_neta['2021'].isnull().sum())
```

```python
# Eliminar filas con valores nulos en la columna '2021'
Mig_neta = Mig_neta.dropna(subset=['2021'])
```

```python
print(Mig_neta['2021'].isnull().sum())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.impute import SimpleImputer


# Seleccionar las columnas de años desde 1960 hasta 2020 como características
features = Mig_neta.columns[3:-1]

# Separar los datos en características (X) y variable objetivo (y)
X = Mig_neta[features]
y = Mig_neta['2021']  # Columna correspondiente a 2021

# Manejar los valores faltantes en X
imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X)

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X_imputed, y, test_size=0.2, random_state=42)

# Crear el modelo de regresión lineal
model = LinearRegression()

# Entrenar el modelo
model.fit(X_train, y_train)

# Realizar predicciones en el conjunto de prueba
y_pred = model.predict(X_test)

# Evaluar el rendimiento del modelo
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse:.2f}')
print(f'R-squared: {r2:.2f}')
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.impute import SimpleImputer


# Seleccionar las columnas de años desde 1960 hasta 2020 como características
features = Mig_neta.columns[3:-1]

# Separar los datos en características (X) y variable objetivo (y)
X = Mig_neta[features]
y = Mig_neta['2021']  # Columna correspondiente a 2021

# Manejar los valores faltantes en X
imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X)

# Crear el modelo de regresión lineal
model = LinearRegression()

# Entrenar el modelo
model.fit(X_imputed, y)

# Predecir los valores de migración neta para el año 2022
X_pred_2022 = X_imputed  # Usar todos los datos imputados para predecir 2022
y_pred_2022 = model.predict(X_pred_2022)

# Crear un DataFrame con las predicciones para el año 2022
predictions_2022 = pd.DataFrame({
    'Country Name': Mig_neta['Country Name'],
    'Country Code': Mig_neta['Country Code'],
    '2022': y_pred_2022
})

# Imprimir el DataFrame con las predicciones para el año 2022
predictions_2022. head(10)
```

Regresión Ridge

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.impute import SimpleImputer



# Seleccionar las columnas de años desde 1960 hasta 2020 como características
features = Mig_neta.columns[3:-1]

# Separar los datos en características (X) y variable objetivo (y)
X = Mig_neta[features]
y = Mig_neta['2021']  # Columna correspondiente a 2021

# Manejar los valores faltantes en X
imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X)

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X_imputed, y, test_size=0.2, random_state=42)

# Crear el modelo de regresión Ridge
ridge_model = Ridge(alpha=1.0)  #  ajustar el valor de alpha según sea necesario

# Entrenar el modelo
ridge_model.fit(X_train, y_train)

# Predecir los valores de migración neta para el año 2022
X_pred_2022 = X_imputed  # Usar todos los datos imputados para predecir 2022
y_pred_2022 = ridge_model.predict(X_pred_2022)

# Calcular las métricas de rendimiento
mse = mean_squared_error(y_test, ridge_model.predict(X_test))
r2 = r2_score(y_test, ridge_model.predict(X_test))

# Crear un DataFrame con las predicciones para el año 2022
predictions_2022 = pd.DataFrame({
    'Country Name': Mig_neta['Country Name'],
    'Country Code': Mig_neta['Country Code'],
    '2022': y_pred_2022
})


predictions_2022

# Imprimir las métricas de rendimiento
print(f'Mean Squared Error: {mse:.2f}')
print(f'R-squared: {r2:.2f}')
```

RandomForestRegressor

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.impute import SimpleImputer



# Seleccionar las columnas de años desde 1960 hasta 2020 como características
features = Mig_neta.columns[3:-1]

# Separar los datos en características (X) y variable objetivo (y)
X = Mig_neta[features]
y = Mig_neta['2021']  # Columna correspondiente a 2021

# Manejar los valores faltantes en X
imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X)

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X_imputed, y, test_size=0.2, random_state=42)

# Crear el modelo de Random Forest Regressor
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)  # ajustar los hiperparámetros según sea necesario

# Entrenar el modelo
rf_model.fit(X_train, y_train)

# Predecir los valores de migración neta para el año 2022
X_pred_2022 = X_imputed  # Usar todos los datos imputados para predecir 2022
y_pred_2022 = rf_model.predict(X_pred_2022)

# Calcular las métricas de rendimiento
mse = mean_squared_error(y_test, rf_model.predict(X_test))
r2 = r2_score(y_test, rf_model.predict(X_test))

# Crear un DataFrame con las predicciones para el año 2022
predictions_2022 = pd.DataFrame({
    'Country Name': Mig_neta['Country Name'],
    'Country Code': Mig_neta['Country Code'],
    '2022': y_pred_2022
})


# Imprimir las métricas de rendimiento
print(f'Mean Squared Error: {mse:.2f}')
print(f'R-squared: {r2:.2f}')
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.impute import SimpleImputer



# Seleccionar las columnas de años desde 1960 hasta 2020 como características
features = Mig_neta.columns[3:-1]

# Separar los datos en características (X) y variable objetivo (y)
X = Mig_neta[features]
y = Mig_neta['2021']  # Columna correspondiente a 2021

# Manejar los valores faltantes en X
imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X)

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X_imputed, y, test_size=0.2, random_state=42)

# Crear el modelo de KNeighborsRegressor
k_neighbors_model = KNeighborsRegressor(n_neighbors=5)  #  ajustar el número de vecinos según sea necesario

# Entrenar el modelo
k_neighbors_model.fit(X_train, y_train)

# Predecir los valores de migración neta para el año 2022
X_pred_2022 = X_imputed  # Usar todos los datos imputados para predecir 2022
y_pred_2022 = k_neighbors_model.predict(X_pred_2022)

# Calcular las métricas de rendimiento
mse = mean_squared_error(y_test, k_neighbors_model.predict(X_test))
r2 = r2_score(y_test, k_neighbors_model.predict(X_test))

# Crear un DataFrame con las predicciones para el año 2022
predictions_2022 = pd.DataFrame({
    'Country Name': Mig_neta['Country Name'],
    'Country Code': Mig_neta['Country Code'],
    '2022': y_pred_2022
})



# Imprimir las métricas de rendimiento
print(f'Mean Squared Error: {mse:.2f}')
print(f'R-squared: {r2:.2f}')
```
