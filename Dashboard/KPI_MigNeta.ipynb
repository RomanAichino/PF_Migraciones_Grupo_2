```python
import pandas as pd
```

```python
df = pd.read_excel('NetMigration.xlsx')
```

```python
df.head(10)
```

```python
# Crear un nuevo DataFrame solo para Argentina
df_argentina = df.loc[df['Country Name'] == 'Argentina']

# Mostrar el nuevo DataFrame
df_argentina
```

```python
# Crear un nuevo DataFrame solo para Argentina y reiniciar el índice
df_argentina = df[df['Country Name'] == 'Argentina'].reset_index(drop=True)

# Mostrar el nuevo DataFrame
df_argentina.head(10)
```

```python
# Calcular el porcentaje de variación en la columna "Migración Neta"
df_argentina['Variación %'] = df_argentina['Migración Neta'].pct_change()

# Mostrar el DataFrame resultante
df_argentina
```

```python
# Calcular el incremento del 1% respecto del año anterior
df_argentina['Migración Neta Incrementado'] = df_argentina['Migración Neta'].shift(1) * 1.01

# Mostrar el DataFrame resultante
df_argentina
```

```python
# Eliminar la fila correspondiente a 2022 en la columna "Year"
year_to_drop = 2022
df_argentina = df_argentina[df_argentina['Year'] != year_to_drop]
```

```python
df_argentina
```

```python
# Guardar el dataframe como archivo CSV
df_argentina.to_csv('NetMigrationArg.csv', index=False)
```
