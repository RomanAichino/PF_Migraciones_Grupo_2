## **Migración neta**
La migración neta es el total neto de personas que migraron durante el período: la cantidad total de inmigrantes menos la cantidad anual de emigrantes, incluidos los ciudadanos y los no ciudadanos. Los datos son estimaciones de cinco años. Para derivar estimaciones de migración neta, la División de Población de las Naciones Unidas toma en cuenta la historia de migraciones del pasado de un país o zona, la política de migraciones del país y el ingreso de refugiados en períodos recientes.  Los datos para calcular estas estimaciones oficiales provienen de una variedad de fuentes, incluidas las estadísticas fronterizas, los registros administrativos, las encuestas y los censos.  Cuando no se pueden realizar estimaciones oficiales porque los datos son insuficientes, se  deriva la migración neta de la ecuación de balance, que es la diferencia entre el crecimiento general de la población y el aumento natural durante el período entre censos de 1990 a 2000.

```python
# Importamos el dataset de la migración neta en el continente Americano 
Migracion_Neta_df = pd.read_csv("Data/MigracionNetaAñoPais1.csv")
# Migracion_Neta_df.head(3)

# Seleccionamos los datos relacionadoa a Argentina en un nuevo dataframe
Migracion_Neta_Argentina_df = Migracion_Neta_df[['Año','Argentina']]
# Migracion_Neta_Argentina_df

# Graficamos 
sns.lineplot(data=Migracion_Neta_Argentina_df, x=Migracion_Neta_Argentina_df['Año'], y=Migracion_Neta_Argentina_df['Argentina'])
plt.title('Migración Neta Argentina de 1960 a 2021')
plt.ylabel('Migración Neta')
plt.xticks(rotation=90);
```

Notamos que entre 1960 y 1970 Argentina tuvo una migración neta positiva más alta por lo que hubo más flujo de inmigrantes que de migrantes, y sus valores mas bajos entre 1995 y 2005 por lo cual hubo mayor cantidad de migrantes que de inmigrantes

## **GDP per capita (current LCU)**
El PIB per cápita es el producto interno bruto dividido por la población a mitad de año. El PIB es la suma del valor agregado bruto de todos los productores residentes en la economía más los impuestos sobre los productos y menos los subsidios no incluidos en el valor de los productos. Se calcula sin hacer deducciones por depreciación de bienes fabricados o por agotamiento y degradación de los recursos naturales. Los datos están en la moneda local actual.

```python
GDP_per_capita_LCU_df = pd.read_excel("Data_ETL/GDP_per_capita_LCU.xlsx")
GDP_per_capita_LCU_df.head(3)
```

```python
# Filtramos los datos de Argentina
GDP_per_capita_LCU_Argentina_df = GDP_per_capita_LCU_df[GDP_per_capita_LCU_df['Country Name']=='Argentina']
GDP_per_capita_LCU_Argentina_df = GDP_per_capita_LCU_Argentina_df.reset_index(drop=True)
GDP_per_capita_LCU_Argentina_df.head()
```

```python
# Graficamos 
plt.figure(figsize=(15, 5))
sns.lineplot(data=GDP_per_capita_LCU_Argentina_df, x=GDP_per_capita_LCU_Argentina_df['Year'], y=GDP_per_capita_LCU_Argentina_df['GDP_per_capita_LCU'])
plt.title('GDP_per_capita_LCU')
plt.ylabel('')
plt.xticks(rotation=90);
```

## **Personal remittances, received (% of GDP)**
Las remesas personales comprenden las transferencias personales y la remuneración de los empleados. Las transferencias personales consisten en todas las transferencias corrientes en efectivo o en especie realizadas o recibidas por hogares residentes hacia o desde hogares no residentes. Las transferencias personales incluyen, por lo tanto, todas las transferencias corrientes entre personas físicas residentes y no residentes. La remuneración de los empleados se refiere a los ingresos de los trabajadores fronterizos, de temporada y otros trabajadores a corto plazo que están empleados en una economía en la que no son residentes y de los residentes empleados por entidades no residentes. Los datos son la suma de dos elementos definidos en la sexta edición del Manual de Balanza de Pagos del FMI: transferencias personales y remuneración de los empleados.

```python
Personal_remittances_received_df = pd.read_excel("Data_ETL/Personal_remittances_received_df.xlsx")
Personal_remittances_received_df.head(3)   
```

```python
# Filtramos los datos de Argentina
Personal_remittances_received_df = Personal_remittances_received_df[Personal_remittances_received_df['Country Name']=='Argentina']
Personal_remittances_received_df = Personal_remittances_received_df.reset_index(drop=True)
Personal_remittances_received_df.head()
```

```python

# Graficamos 
plt.figure(figsize=(15, 5))
sns.lineplot(data=Personal_remittances_received_df, x=Personal_remittances_received_df['Year'], y=Personal_remittances_received_df['Personal_remittances_received'])
plt.title('Personal_remittances_received')
plt.ylabel('')
plt.xticks(rotation=90);
```
