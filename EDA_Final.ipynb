## **Migración neta**
La migración neta es el total neto de personas que migraron durante el período: la cantidad total de inmigrantes menos la cantidad anual de emigrantes, incluidos los ciudadanos y los no ciudadanos. Los datos son estimaciones de cinco años. Para derivar estimaciones de migración neta, la División de Población de las Naciones Unidas toma en cuenta la historia de migraciones del pasado de un país o zona, la política de migraciones del país y el ingreso de refugiados en períodos recientes.  Los datos para calcular estas estimaciones oficiales provienen de una variedad de fuentes, incluidas las estadísticas fronterizas, los registros administrativos, las encuestas y los censos.  Cuando no se pueden realizar estimaciones oficiales porque los datos son insuficientes, se  deriva la migración neta de la ecuación de balance, que es la diferencia entre el crecimiento general de la población y el aumento natural durante el período entre censos de 1990 a 2000.

```python
# Importamos el dataset de la migración neta en el continente Americano 
Migracion_Neta_df = pd.read_csv("Data/MigracionNetaAñoPais1.csv")
# Migracion_Neta_df.head(3)

# Seleccionamos los datos relacionadoa a Argentina en un nuevo dataframe
Migracion_Neta_Argentina_df = Migracion_Neta_df[['Año','Argentina']]
# Migracion_Neta_Argentina_df

# Graficamos 
sns.lineplot(data=Migracion_Neta_Argentina_df, x=Migracion_Neta_Argentina_df['Año'], y=Migracion_Neta_Argentina_df['Argentina'])
plt.title('Migración Neta Argentina de 1960 a 2021')
plt.ylabel('Migración Neta')
plt.xticks(rotation=90);
```

Notamos que entre 1960 y 1970 Argentina tuvo una migración neta positiva más alta por lo que hubo más flujo de inmigrantes que de migrantes, y sus valores mas bajos entre 1995 y 2005 por lo cual hubo mayor cantidad de migrantes que de inmigrantes

## **Emigración**

```python
Emigracion_Argentina_df = pd.read_csv("Data/EmigracionArgentina.csv")
#Emigracion_Argentina_df .head(3)

Emigracion_Argentina_Total_df = Emigracion_Argentina_df[Emigracion_Argentina_df['Migration by Gender Name']=='Total']
Emigracion_Argentina_Total_df.head(3)

```

```python
df = Emigracion_Argentina_Total_df[['Country Dest Name','No. Migrantes']]
df.groupby(by = 'Country Dest Name').sum()
df = df.sort_values(by='No. Migrantes',ascending=True)
```

```python
sns.barplot(x = 'Country Dest Name', y = 'No. Migrantes', data = df)
plt.title('Emigración de Argentina a Otros Paises en América')
plt.xticks(rotation=90);
```

Como se observa en el gráfico de barras hay mayor emigración de argentinos a **Estados Unidos de América** y **Paraguay**

## **Inmigración**

```python
Inmigracion_Argentina_df = pd.read_excel("Data/InmigracionArgentina.xlsx")
#Inmigracion_Argentina_df .head(3)

Inmigracion_Argentina_Total_df = Inmigracion_Argentina_df[Inmigracion_Argentina_df['Migration by Gender Name']=='Total']
Inmigracion_Argentina_Total_df.head(3)

```

```python
df = Inmigracion_Argentina_Total_df[['Country Origin Name','Inmigrantes']]
df.groupby(by = 'Country Origin Name').sum()
df = df.sort_values(by='Inmigrantes',ascending=True)
```

```python
sns.barplot(x = 'Country Origin Name', y = 'Inmigrantes', data = df)
plt.title('Inmigración de Paises en América a Argentina')
plt.xticks(rotation=90);
```

Como vemos en la gráfica, **Paraguay**, **Chile** y **Bolivia** son los países con más inmigrantes a Argentina
