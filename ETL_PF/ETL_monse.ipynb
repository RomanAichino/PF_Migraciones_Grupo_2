```python
import pandas as pd
import numpy as np
```

## **Unemployment, total (% of total labor force) (national estimate)**

```python
# Cargamos, eliminamos duplicados y reseteamos el indice
DesempleoTotal_df = pd.read_csv("Data/datasets_desempleo_total.csv")
DesempleoTotal_df.drop_duplicates(inplace=True)
DesempleoTotal_df = DesempleoTotal_df.reset_index(drop=True)
DesempleoTotal_df.head(3)
```

```python
DesempleoTotal_df = pd.melt(DesempleoTotal_df,id_vars=['Country Name','Country Code','Indicator Name','Indicator Code'], var_name="Year", value_name="Unemployment percentage of total labor force")
DesempleoTotal_df.head(5)
```

```python
DesempleoTotal_df['id'] = DesempleoTotal_df['Country Code'].str.cat(DesempleoTotal_df['Year'])
DesempleoTotal_df.head(5)
```

```python
# Eliminamos columnas que no serÃ¡n usadas 
DesempleoTotal_df = DesempleoTotal_df.drop(columns=['Indicator Name','Indicator Code']).copy()
```

```python
DesempleoTotal_df = DesempleoTotal_df.reset_index(drop=True)
```

```python
DesempleoTotal_df.head(5)
```

```python
#DesempleoTotal_df.to_excel("Data_ETL/DesempleoTotal.xlsx",index=False)
```
