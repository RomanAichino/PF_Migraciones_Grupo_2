## **GDP per capita (current LCU)**

```python
GDP_per_capita_LCU_df['id'] = GDP_per_capita_LCU_df['Country Code'].str.cat(GDP_per_capita_LCU_df['Year'])
GDP_per_capita_LCU_df.head(5)
```

```python
TasaPobreza_df = TasaPobreza_df.drop(columns=['Indicator Name','Indicator Code']).copy()
TasaPobreza_df = TasaPobreza_df.reset_index(drop=True)
```
